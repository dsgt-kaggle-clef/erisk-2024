```python
df_res_2023['USER'] = df_res_2023['USER'].str.replace('erisk2023-T3_Subject', 'Subject')
df_emb_2023['USER'] = df_emb_2023['USER'].str.replace('eRisk2022-T3_Subject', 'Subject')

df_res_2022['USER'] = df_res_2022['USER'].str.replace('erisk2022-T3_Subject', 'Subject')
df_emb_2022['USER'] = df_emb_2022['USER'].str.replace('eRisk2022-T3_Subject', 'Subject')
```

```python
df_res_2023
```

```python
# combining 2022 and 2023 training data
df_res_2023['USER'] = ['Subject' + str(i) for i in range(len(df_res_2022) + 1, len(df_res_2022) + 1 + len(df_res_2023))]

df_res = pd.concat([df_res_2022, df_res_2023], ignore_index=True)

df_res
```

```python
max_subject_2022 = 28
```

```python
df_emb_2023['USER'] = df_emb_2023['USER'].str.replace('Subject', '').astype(int) + max_subject_2022
df_emb_2023['USER'] = 'Subject' + df_emb_2023['USER'].astype(str)
```

```python
df_emb_2023
```

```python
df_emb = pd.concat([df_emb_2022, df_emb_2023], ignore_index=True)
```

```python
df_emb
```

```python
df_emb_mean = df_emb.groupby('USER').mean().reset_index()
```

```python
df_emb_mean
```

```python
# merging USER responses and text embedding for each user
df_merged = pd.merge(df_res, df_emb_mean, on='USER', how='inner')
```

```python
df_merged
```

```python
# dimensionality reduction
features = [col for col in df_merged if col.startswith('embedding_')]
X_embeddings = df_merged[features]
```

```python
# standardising features.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_embeddings)
```

```python
# PCA to reduce dimensions to 50
pca = PCA(n_components=50)  # Example: reduce to 50 dimensions
X_pca = pca.fit_transform(X_scaled)
```

```python
for i in range(X_pca.shape[1]):
    df_merged[f'pca_component_{i}'] = X_pca[:, i]
```

```python
df_merged
```

```python
embedding_columns = [col for col in df_merged if col.startswith('embedding_')]

df_merged.drop(columns=embedding_columns, inplace=True)
```
