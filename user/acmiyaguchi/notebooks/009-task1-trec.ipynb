```python
# pandas read csv directly from gcs
import pandas as pd
import trectools

df = pd.read_csv(
    "gs://dsgt-clef-erisk-2024/task1/processed/eval/nb_count/v3/inference/predictions.csv"
)
df
```

```python
df.sort_values(["symptom_number", "position_in_ranking"]).rename(
    columns=dict(
        # rank="position_in_ranking", query="symptom_number", q0="Qo", docid="sentence_id"
        position_in_ranking="rank",
        symptom_number="query",
        Qo="q0",
        sentence_id="docid",
        system_name="system",
    )
)
```

```python
from IPython.display import display

# convert the dataframe to trec format
trec = trectools.TrecRun()
trec.load_run_from_dataframe(
    df.sort_values(["symptom_number", "position_in_ranking"]),
    column_mapping=dict(
        # rank="position_in_ranking", query="symptom_number", q0="Qo", docid="sentence_id"
        position_in_ranking="rank",
        symptom_number="query",
        Qo="q0",
        sentence_id="docid",
        system_name="system",
    ),
)
# display(trec.run_data)

# write to trec format xml
trec.export_document_list(filename="mypool.txt", with_format="relevation")
```
