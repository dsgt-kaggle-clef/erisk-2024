```python
import xml.etree.ElementTree as ET

import pandas as pd
import nltk
import string
import re
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
import xml.etree.ElementTree as et
import numpy as np
```

```python
def read_posts(writing):
    posts = []
    for wrt in writing:
        user_post = {}
        user_post['TITLE'] = wrt.find('TITLE').text
        user_post['DATE'] = wrt.find('DATE').text
        user_post['INFO'] = wrt.find('INFO').text
        user_post['TEXT'] = wrt.find('TEXT').text
        posts.append(user_post)
        training_sample = pd.DataFrame(posts)
    return training_sample
```

```python
def read_posts_2023(writing):
    posts = []
    #  the 2023 user posts have text in byte string format. Skip initial characters and later chars
    for wrt in writing:
        user_post = {}
        user_post['TITLE'] = wrt.find('TITLE').text[3:-2]
        user_post['DATE'] = wrt.find('DATE').text[3:-2]
        user_post['INFO'] = wrt.find('INFO').text
        user_post['TEXT'] = wrt.find('TEXT').text[3:-2]
        posts.append(user_post)
        training_sample = pd.DataFrame(posts)
    return training_sample
```

```python
from bs4 import BeautifulSoup

data_2022 = pd.DataFrame([])
for i in range(1, 29):
    with open(f"/Users/aaryanpotdar/Desktop/Kaggle_CLEF/task3/training/t3_training/TRAINING DATA (FROM ERISK 2022 AND 2023)/2022/T3 2022/eRisk2022_T3_Collection/eRisk2022-T3_Subject{i}.xml") as fp:
        soup = BeautifulSoup(fp, 'xml')
    writing = soup.find_all('WRITING')
    user_sample = read_posts(writing)
    user_sample['USER'] = f'eRisk2022-T3_Subject{i}'
    data_2022 = pd.concat([data_2022, user_sample])
```

```python
# removing words begining with r/ or u/. These are specific to reddit and don't add meaning to context
def remove_r_u(text):
    pattern = r'\b(?:u/|r/)\w+\b'
    return re.sub(pattern, '', text)
```
